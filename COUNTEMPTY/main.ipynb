```python
import os
import pandas as pd
import matplotlib.pyplot as plt
path = r"Data"
```

```python
folder = [file for file in os.listdir(path)]
df = pd.concat([pd.read_excel(path+'\\'+file) for file in folder])
df.head()
df.columns = df.columns.str.replace(' ', '_')
```

```python
emp = []
for col in df.columns:
    emp.append(df[col].isna().sum())
emp
```

```python
ind = ['Storage Type', 'Storage Bin', 'Posting Date',
 'Handling Unit', 'Product', 'Product Short Description', 'Quantity']
plt.style.use('ggplot')
plt.title('Empty Rows for Octomber')
plt.xlabel('Columns Names')
plt.ylabel('Empty Rows')
plt.xticks(rotation = 90)
plt.bar(ind, emp)
plt.show()
```

```python
labels = ind[2:]
sizes = emp[2:]
explode = ( 0.1, 0, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.title('Ration Empty Rows Octomber')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

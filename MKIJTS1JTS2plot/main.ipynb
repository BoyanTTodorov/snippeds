```python
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
import os
import datetime

path_db = r"S:\Warehouse dept\Inventory control\Reporting and dashboards\Bin's Counting\Cycle Counting\DATABASE\cycle_couting.db"
```

```python
conn = sqlite3.connect(path_db)
query = "SELECT * FROM MKI"
mki = pd.read_sql(query, conn)
```

```python
data = []
week = datetime.datetime.utcnow().isocalendar()[1]
print(week)
for i in range(1,week+1):
    data.append(mki[str(i)].astype(int).sum())
plt.title("MKI Counting's by Week")
plt.plot(range(1,week+1), data)
```

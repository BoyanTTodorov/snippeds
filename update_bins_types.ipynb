```python
import pandas as pd
import sqlite3

path_bins = r"C:\Users\b.todorov\Desktop\Blocked Bins\B500570.XLSX"
path_db = r"C:\Users\b.todorov\Desktop\Bin's Counting\Cycle Counting\DATABASE\cycle_couting.db"
df = pd.read_excel(path_bins)

conn = sqlite3.connect(path_db)
curs = conn.cursor()
for row in df.itertuples():
    curs.execute(f'''INSERT INTO JTS2 ("Storage Bin", "Type") VALUES ("{row[1]}", "{row[2]}")''')
    print(f'Updating Bin: {row[1]} with type: {row[2]}')
conn.commit()
```
